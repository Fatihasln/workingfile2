## 3: Plot number of age of killed people under two groups : Under 25 and Above 25

```python
df.head()
```

```python

kill["age_cat"] = df.age.apply(lambda x : "above25" if x>= 25 else "below25")
```

```python
kill.head()
```

```python
age_class  = kill["age_cat"].value_counts()
```

```python
# with matplotlib

fig, ax = plt.subplots()
ax.bar(x= age_class.index, height=age_class.values);


for p in ax.patches:
    ax.annotate(round(p.get_height()), (p.get_x()+0.35, p.get_height()+0.25));
```

**INSTRUCTION-2: Demonstrate Race Ratio of Killed People by a Pie Chart (You can use Matplotlib)**

```python
df.race.value_counts()
```

```python
label = df.race.value_counts().index
```

```python
values = df.race.value_counts().values
```

```python
# matplotlib

explode = (0,0,0.1,0,0,0)

fig,ax= plt.subplots(figsize= (16,10))

ax.pie(x = values,
        explode=explode,
        labels=label,
        labeldistance=0.8,
        autopct="%.1f%%",
        startangle=90,
        textprops={"color": "darkred"});
```
